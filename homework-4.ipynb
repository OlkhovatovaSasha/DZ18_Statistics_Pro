**Написать линейную регрессию на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT.**



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
MEDV_1 = np.array(MEDV).reshape(-1,1)
LSTAT_1 = np.array(LSTAT).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    MEDV_1,
    LSTAT_1,
    test_size = 0.2
)
```

```python
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
```

```python
print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('MEDV_1')
plt.ylabel('LSTAT_1')
plt.title('Линейная регрессия от одной переменной')
# Построим прямую для регрессии
# Создадим 100 точек для предсказаний
x = np.linspace(-50, 200, 100)
# Посчитаем предсказанные значения
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0,55)
plt.ylim(0,45)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

В данном случае получается зависимость двух переменных: **чем больше MEDV, тем меньше LSTAT и наоборот соответсвенно.**

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
values = numpy.random.randint(1, 7, size=10)
plt.hist(values)

plt.xlabel('Цифра на грани кубика')
plt.ylabel('Количества раз')
plt.title('Подбрасывание кубика 10 раз')
plt.show()
```

Это какое-то разное распределение, смотря какие числа и сколько раз выберет рандом. По логике вещей распеределение должно быть равномерным, с увелечением количества подбрасываний распределение будет равномерным.

Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
number_of_sums = 1000
values_100 = numpy.random.randint(1, 7, size=100)
X1 = np.zeros([number_of_sums])
for i in range(number_of_sums):
    X1[i] = np.sum(numpy.random.randint(1, 7, size=100))/100
plt.hist(X1, 50)
plt.xlabel('X values')
plt.ylabel('Number of elements in each X')
plt.show()
```

Как я написала и выше график изменился из-за того что выборок стало достаточно для оценки распределения. Так же поэтому можно применять ЦПТ.
